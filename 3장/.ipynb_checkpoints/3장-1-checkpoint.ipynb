```python
# 정규식으로 태그 구조에 쉽게 접근 가능
```

```python
import re 
tmp_string = tmp_one.find(class_='sammyListing').get_text()

re.split(('\n|\r\n'), tmp_string)

print(re.split(('\n|\r\n'), tmp_string)[0])
print(re.split(('\n|\r\n'), tmp_string)[1])
```

```python
from urllib.parse import urljoin

rank = []
main_menu = []
cafe_name = []
url_add = []

list_soup = soup.find_all('div', 'sammy') # 태그의 sammy클래스 찾아서 저장

for item in list_soup:
    rank.append(item.find(class_='sammyRank').get_text()) # 순위 텍스트 저장
    
    tmp_string = item.find(class_='sammyListing').get_text() # 텍스트 저장

    main_menu.append(re.split(('\n|\r\n'), tmp_string)[0]) # 보기 쉽게 정규식 사용
    cafe_name.append(re.split(('\n|\r\n'), tmp_string)[1])
    
    url_add.append(urljoin(url_base, item.find('a')['href'])) # 링크 찾아 저장
    # 모든 태그의 경로가 같지 않을 수 있다. urljoin은 절대경로로 통일하기 위해 사용
```

```python
print(rank[:5])
print('--------------------------------------------------------------------------------------------')
print(main_menu[:5])
print('--------------------------------------------------------------------------------------------')
print(cafe_name[:5])
print('--------------------------------------------------------------------------------------------')
print(url_add[:5])
```

```python
len(rank), len(main_menu), len(cafe_name), len(url_add) # 50개 다 잘받아왔음
```

```python
import pandas as pd

data = {'Rank':rank, 'Menu':main_menu, 'Cafe':cafe_name, 'URL':url_add}
df = pd.DataFrame(data) # 딕셔너리 -> 데이터 프레임
df.head()
```

```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen

import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\user\Desktop\사용파일\데이터분석\데이터 분석\DATA\파이썬으로 데이터 주물럭\DataScience-master\data\03. best_sandwiches_list_chicago.csv', index_col=0)
df.head()
```

```python
df['URL'][0]
```

```python
import ssl
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

context = ssl._create_unverified_context()

req = Request(df['URL'][0], headers={'User-Agent': 'Mozilla/5.0'})
html = urlopen(req, context=context)
soup_tmp = bs(html, "html.parser")
soup_tmp
```

```python
from tqdm import tqdm_notebook

price = []
address = []

for n in tqdm_notebook(df.index):
    context = ssl._create_unverified_context()

    req = Request(df['URL'][n], headers={'User-Agent': 'Mozilla/5.0'})

    html = urlopen(req, context=context)
    soup_tmp = bs(html, 'lxml')
    
    gettings = soup_tmp.find('p', 'addy').get_text()
    
    price.append(gettings.split()[0][:-1])
    address.append(' '.join(gettings.split()[1:-2]))
```

```python
# 3 - 7 50개 웹 페이지에 대한 정보 가져오기
```

```python
price
```

```python
address
```

```python
# 위 adress에 multiple이라는 값이 있음
```

```python
len(price), len(address), len(df)
```

```python
df['Price'] = price
df['Address'] = address

df = df.loc[:, ['Rank', 'Cafe', 'Menu', 'Price', 'Address']]
df.set_index('Rank', inplace=True)
df.head()
```

```python
import folium
import pandas as pd
import googlemaps
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\user\Desktop\사용파일\데이터분석\데이터 분석\DATA\파이썬으로 데이터 주물럭\DataScience-master\data\03. best_sandwiches_list_chicago2.csv', index_col=0)
df.head()
```

```python
import googlemaps
```

```python
gmaps_key = "*********api키" 
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
import numpy as np
lat = []
lng = []

for n in tqdm_notebook(df.index):
    if df['Address'][n] != 'Multiple': # 멀티플 이면 뺴줌
        target_name = df['Address'][n]+', '+'Chicago' 
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
        
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
len(lat), len(lng)
```

```python
df['lat'] = lat
df['lng'] = lng
df.head()
```

```python
import folium
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], 
                                      zoom_start=11)
folium.Marker([df['lat'].mean(), df['lng'].mean()], 
                                      popup='center').add_to(mapping)
mapping
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], 
                     zoom_start=11)

for n in df.index:
    if df['Address'][n] != 'Multiple':
        folium.Marker([df['lat'][n], df['lng'][n]], 
                                      popup=df['Cafe'][n]).add_to(mapping)

mapping
```
